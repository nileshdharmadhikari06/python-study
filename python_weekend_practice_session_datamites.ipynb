```python
name= "Nilesh"

#forword endexing (left to right)      starting from 0 first letter will be count as 0
#backword indexing (right to left)     starting from -1 first letter from right will be count as -1

print(name[3])
print(name[-1])

# SLICING

print(name[::-1])
print(name[-6:-1])
print(name[:])
print(name[0:6])
```
