```python
def find_repeating_letters(word):
    word = word.lower()  # make it case-insensitive
    seen = set()
    repeats = set()

    for letter in word:
        if letter in seen:
            repeats.add(letter)
        else:
            seen.add(letter)

    return repeats
```
