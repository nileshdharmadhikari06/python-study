```python
# Cell 1: Define a function to check prime numbers from 1 to 100.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
primes = [num for num in range(1, 101) if is_prime(num)]
print("Prime numbers between 1 and 100 are:")
print(primes)
```

```python
inventory = {
    "apple": 50,
    "banana": 20,
    "orange": 54,
    "mango": 15
}

def add_stock(item, quantity):
    """Add stock for an item."""
    if item in inventory:
        inventory[item] += quantity
    else:
        inventory[item] = quantity

def sell_item(item, quantity):
    """Sell items without exceeding available stock."""
    if item not in inventory:
        print(f"{item} not found in inventory.")
        return
    if quantity > inventory[item]:
        print(f"Cannot sell {quantity} {item}(s). Only {inventory[item]} available.")
    else:
        inventory[item] -= quantity
        print(f"Sold {quantity} {item}(s). Remaining: {inventory[item]}")

def out_of_stock_items():
    """Return a list of items that are out of stock."""
    return [item for item, qty in inventory.items() if qty == 0]

def sorted_inventory():
    """Return inventory sorted by quantity (highest first)."""
    return dict(sorted(inventory.items(), key=lambda x: x[1], reverse=True))
```

```python
add_stock("apple", 10)
sell_item("banana", 4)
print("Out of stock:", out_of_stock_items())
print("Sorted inventory:", sorted_inventory())
```
