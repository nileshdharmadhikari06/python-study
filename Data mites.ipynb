```python
def outlier(x):
    out=[]
    q1=np.quantile(x,.25)
    q3=np.quantile(x,.75)

    iqr=q3-q1
    upper=q3+(1.5*iqr)
    lower=q1-(1.5*iqr)

    for num in x:
        if num<lower or num>upper:
            out.append(num)
    return out
```

```python
outlier(x)
```

```python
x=[1,2,3,4,5,6,7,8,46]
def remove_outlier(x):
    out=[]
    q1=np.quantile(x,.25)
    q3=np.quantile(x,.75)

    iqr=q3-q1
    upper=q3+(1.5*iqr)
    lower=q1-(1.5*iqr)

    for num in x:
        if num>lower and num<upper:
            out.append(num)
    return out
```

# Hypothesis

```python
import numpy as np
from statsmodels.stats.weightstats import ztest
import scipy.stats as stats

# given data
pop_mean = 400
pop_std = 25
sample_mean = 250
sample_size = 50
alpha = 0.05



Z_statistic = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
print(Z_statistic)

p_value = 2 * stats.norm.cdf(Z_statistic)

print(f"Z_statistic: {Z_statistic:.5f}")
print(f"p-value: {p_value:.5f}")

if p_value < alpha:
    print("Decision: Reject the null hypothesis (H0)")
else:
    print("Decision: Fail to reject the null hypothesis (H0)")

if p_value < alpha:
    print(f"conclusion: there is sufficient evidence to say that the average monthly bill is significantly different from the service provider's claim of {pop_mean}.")
else:
    print(f"conclusion: there is not enough evedince to say that the average monthly bill is different from the service providers claim {pop_mean}.")
```

```python
import numpy as np
from scipy import stats

b = np.array([120,122,118,130,125,128,115,121,123,119])
a = np.array([125,120,112,128,122,125,110,117,119,114])

alpha = 0.05

t_stats, p_val = stats.ttest_rel(a,b)

m = np.mean(a-b)
s = np.std(a-b, ddof=1)
n = len(b)
t_manual = m / (s/np.sqrt(n))

decision = "Reject" if p_val <=alpha else "Accept null hyphothesis"
concl = "Significant difference." if decision == "Reject" else "No significant difference."

print("T:", t_stats)
print("p:", p_val)
print("T_manual:", t_manual)
print(f"Decision: {decision} H0 at ={alpha}")
print("Conclusion", concl)
```

```python
import numpy as np
group1=np.array([85,90,88,75,78])
group2=np.array([80,82,84,79,81])
group3=np.array([90,92,94,89,93])

from scipy import stats

f_stat,p_value=stats.f_oneway(group1,group2,group3)

print("F-statistics:",f_stat)
print("p-value :",p_value)

if p_value<0.5:
    print("reject null hypothesis")
else:
    print("Accept null hyphothesis")
```
