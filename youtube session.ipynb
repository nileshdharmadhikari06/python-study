```python
import matplotlib.pyplot as plt

# Function to calculate digit sum
def calcSumofDigits(x):
    digit4 = x % 10          # Ones
    digit3 = (x // 10) % 10   # Tens
    digit2 = (x // 100) % 10  # Hundreds
    digit1 = (x // 1000) % 10 # Thousands
    #print(digit1,digit2,digit3,digit4)
    sum = digit1+digit2+digit3+digit4
    #return sum
    if (sum)>9:
       return int(sum%10) + (int(sum/10)%10) 
    else:
        return sum

# Find numbers where sum of digits is 8
sum_8_numbers = [i for i in range(1000, 10000) if calcSumofDigits(i) == 8]

# Group by thousand series
series_counts = {}
for i in range(1, 10):
    label = f"{i}000s"
    count = len([n for n in sum_8_numbers if i*1000 <= n < (i+1)*1000])
    series_counts[label] = count

# Plotting the graph
plt.bar(series_counts.keys(), series_counts.values(), color='skyblue', edgecolor='navy')
plt.xlabel('Thousand Series')
plt.ylabel('Count of Numbers (Sum of Digits = 8)')
plt.title('Numbers with Digit Sum of 8 per 1000 Series')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.savefig('digit_sum_distribution.png')
```
